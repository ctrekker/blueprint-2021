```python
import keras
import numpy as np
import librosa
import librosa.display
import librosa.beat
import soundfile as sf
from matplotlib import pyplot as plt
```

```python
dataset_root = './noise_dataset'
```

```python
sample_path = dataset_root + '/clean/common_voice_en_1.wav'
```

```python
y, sr = librosa.load(sample_path)
```

```python
plt.plot(y)
```

```python
librosa.get_duration(y)
```

```python
librosa.get_samplerate(sample_path)
```

Above means the sample rate is 48KHZ, which is an indicator for quality. Note this for upscaling

```python
librosa.display.waveplot(y)
```

```python
def plot_spectogram(y):
    fig, ax = plt.subplots(nrows=1, ncols=1, sharex=True)
    D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
    img = librosa.display.specshow(D, y_axis='linear', x_axis='time',
                                   sr=sr, ax=ax)
    ax.set(title='Linear-frequency power spectrogram')
    ax.label_outer()
    
    fig
```

```python
plot_spectogram(y)
```

```python
D.shape
```

```python
librosa.beat.tempo(y)
```

```python
noise_mean = 0.005
noise_std = 0.004
```

```python
y_noisy = y + abs(np.random.normal(noise_mean, noise_std))*np.random.randn(len(y))
plot_spectogram(y_noisy)
```

```python
sf.write('stereo_file.wav', y_noisy, sr, 'PCM_24')
```
